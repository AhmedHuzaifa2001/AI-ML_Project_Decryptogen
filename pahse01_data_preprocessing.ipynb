```python

df = pd.read_csv('personality_dataset.csv')


print("Loaded shape:", df.shape)
df.head()
```

```python
# data types and missing values
print("\n Dataset Info")
df.info()
```

```python
df.select_dtypes(include=['object'])
```

```python
print(df.isnull().sum())
```

```python
## ENCODE TEXT TO NUMBERS FIRST

# Map Yes/No to 1/0 

binary_mapping = {'Yes': 1, 'No': 0}
df['Stage_fear'] = df['Stage_fear'].map(binary_mapping)
df['Drained_after_socializing'] = df['Drained_after_socializing'].map(binary_mapping)
```

```python
# Mapping the Personality to 1/0
target_mapping = {'Extrovert': 1, 'Introvert': 0}
df['Personality'] = df['Personality'].map(target_mapping)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
num_cols = ['Time_spent_Alone', 'Social_event_attendance', 'Going_outside', 'Friends_circle_size', 'Post_frequency']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# used the Median for continuous variables to protect data from being 
# skewed by extreme outliers.
```

```python
binary_cols = ['Stage_fear', 'Drained_after_socializing']
for col in binary_cols:
    df[col] = df[col].fillna(df[col].mode()[0])


# used Mode for the binary categorical variables because they 
# represent discrete classes (Yes/No), and calculating an average or median
#  would result in invalid, fractional categories.
```

```python
# check for null values gain after fillna
df.isnull().sum()
```

```python
df.to_csv('dataset_cleaned.csv', index=False)
```
